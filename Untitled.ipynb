```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sal = pd.read_csv('company_sales_data.csv')
sal.head(4)
```

```python
sal['Order Date'] = pd.to_datetime(sal['Order Date']).dt.date
```

```python
plt.plot("Order Date","Profit",data = sal)
plt.xlabel('Order Date')
plt.ylabel('Profit')
plt.xticks(rotation = 45)
plt.title("Profit by month")
plt.show()
```

```python
plt.plot("Order Date","Profit",data = sal,color = 'red',marker = 'o',linewidth = 0.5 , markersize = 5 )
plt.xlabel("Order Date")
plt.ylabel("Profit")
plt.xticks(rotation = 45)
plt.yticks(rotation = 16)
plt.title("profit by months")
plt.legend(loc= "upper left")
plt.show()
```

```python
mon_sal_pro = sal.groupby(["year month","Product"])["Quantity"].sum().unstack()
mon_sal_pro 
```

```python
plt.figure(figsize=(12, 6))
for product in mon_sal_pro.columns:
    plt.plot(mon_sal_pro.index.astype(str), mon_sal_pro[product], marker="o", label=product)

plt.xlabel("Month")
plt.ylabel("Units Sold")
plt.title("Monthly Sales per Product")
plt.legend(title="Product")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
